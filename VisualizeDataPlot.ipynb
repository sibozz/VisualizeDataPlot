```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("/Users/mawaddi.rachman/Downloads/datasets_hobbies.csv")
clubs = df['Clubs'].tolist()
fans = df['Fans'].tolist()
hatters = df['Hatters'].tolist()
```

```python
print(clubs)
print(fans)
print(hatters)
```

```python
plt.plot(fans, hatters) 
plt.xlabel('x axis horizontal')
plt.ylabel('y axis')
plt.grid()
plt.show()
```

```python
# fans club
plt.bar(clubs, fans)
plt.xlabel("Clubs")
plt.ylabel("Fans")
plt.xticks(rotation=45)
plt.title("scatter plot y vs x")
plt.grid()
```

```python
# hatters club
plt.bar(clubs, hatters)
plt.xlabel("Clubs")
plt.ylabel("Hatters")
plt.xticks(rotation=45)
plt.title("scatter plot y vs x")
plt.grid()
```

```python
plt.scatter(fans, hatters, c=np.random.randint(0, 12, 12), s=np.random.randn(12)*200, data=fans)
plt.xlabel('fans club')
plt.ylabel('hatters club')
plt.show()
```
