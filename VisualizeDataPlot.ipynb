```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/sibozz/VisualizeDataPlot/master/datasets_hobbies.csv")
clubs = df['Clubs'].tolist()
fans = df['Fans'].tolist()
hatters = df['Hatters'].tolist()
```

```python
print(clubs)
print(fans)
print(hatters)
```

```python
fans.sort()
hatters.sort()

plt.plot(fans, hatters) 
plt.xlabel('x axis horizontal')
plt.ylabel('y axis')
plt.grid()
plt.show()
```

```python
# fans club
plt.bar(clubs, fans)
plt.xlabel("Clubs")
plt.ylabel("Fans")
plt.xticks(rotation=45)
plt.title("scatter plot y vs x")
plt.grid()
```

```python
# hatters club
plt.bar(clubs, hatters)
plt.xlabel("Clubs")
plt.ylabel("Hatters")
plt.xticks(rotation=45)
plt.title("scatter plot y vs x")
plt.grid()
```

```python
sz = np.random.randn(12)
sizes = sz * 200
colors = np.random.randint(0, 12, 12)
plt.scatter(fans, hatters, c=colors, s=sizes, data=fans)
plt.xlabel('fans club')
plt.ylabel('hatters club')
plt.show()
print(type(fans))
```

```python
plt.figure(figsize=(12, 3))

plt.subplot(2,3,1)
plt.bar(fans, hatters)
plt.grid()

plt.subplot(1,3,2)
plt.bar(fans, hatters)

plt.subplot(2,3,3)
plt.scatter(fans, hatters)
plt.grid()

plt.subplot(2,3,4)
plt.plot(fans, hatters)

#plt.subplot(2,3,5)
#plt.plot(names, values)

plt.subplot(2,3,6)
plt.barh(fans, hatters)

plt.suptitle('Categorical Plotting')
plt.show()
```

```python
import numpy as np

fans.sort()

pfans = [0.]
for fan in fans:
    x = fan*0.1
    pfans.append(float(x))
    
t = np.arange(0., pfans[-1], 0.2)
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^', t, t**4, 'b^',)
plt.grid()
plt.show()
```

```python
# fans club with multiple figures and axes
def f(t):
    return np.exp(-t)  * np.cos(2*np.pi*t)

fans.sort()
pfans = [0.]
for fan in fans:
    x = fan*0.1
    pfans.append(x)
    
t1 = np.arange(0., pfans[-1], 0.1)
t2 = np.arange(0.0, pfans[-1], 0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
def f1(t):
    return np.exp(-t)

def f2(t):
    return np.cos(2*np.pi*t)

def f3(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

hatters.sort()
phatters = [0.]
for hatter in hatters:
    x = hatter*0.1
    phatters.append(x)

t1 = np.arange(0.0, phatters[-1], 0.08)

plt.figure()

plt.subplot(3,1,1)
plt.plot(t1, f1(t1), 'bo')

plt.subplot(3,1,2)
plt.plot(t1, f2(t1), 'bo')

plt.subplot(3,1,3)
plt.plot(t1, f3(t1), 'bo')
plt.show()
```
